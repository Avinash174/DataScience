```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
classes=['Physics','Biology','Maths','Chemistry','Hindi']
marks=[60,40,55,70,93]
```

```python
plt.pie(marks,labels=classes)
plt.show()
```

```python
color=['red','blue','green','yellow','black']
plt.pie(marks,labels=classes,colors=color)
plt.show()
```

```python
color=['red','blue','green','yellow','purple']
plt.pie(marks,labels=classes,colors=color,autopct='%0.2f%%')
plt.show()
```

```python
expload_value=[0,0,0,0,0.1]
plt.pie(marks,labels=classes,colors=color,autopct='%0.2f%%',explode=expload_value)
plt.show()
```

```python
plt.pie(marks,labels=classes,colors=color,autopct='%0.2f%%',explode=expload_value,shadow=True)
plt.show()
```

```python
plt.pie(marks,labels=classes,colors=color,autopct='%0.2f%%',explode=expload_value,radius=1.6,shadow=True)
plt.show()
```

```python
textProp={'fontsize':'14','color':'k'}
wedgeprops={'linewidth':'3','linestyle':'--','edgecolor':'white'}
plt.pie(marks,labels=classes,colors=color,explode=expload_value,autopct='%0.1f%%',textprops=textProp)
plt.show()
```

```python
textProp={'fontsize':'14','color':'k'}
wedgeprops={'linewidth':'3','linestyle':'--','edgecolor':'white'}
plt.pie(marks,labels=classes,colors=color,explode=expload_value,autopct='%0.1f%%',textprops=textProp,wedgeprops=wedgeprops)
plt.show()
```

```python
textProp={'fontsize':'14','color':'k'}
wedgeprops={'linewidth':'3','linestyle':'--','edgecolor':'white'}
plt.title('Average & Marks Score')
plt.pie(marks,labels=classes,colors=color,explode=expload_value,autopct='%0.1f%%',textprops=textProp,wedgeprops=wedgeprops)
plt.show()
```

```python
textProp={'fontsize':'14','color':'k'}
wedgeprops={'linewidth':'3','linestyle':'--','edgecolor':'white'}
plt.title('Average & Marks Score')
plt.legend()
plt.pie(marks,labels=classes,colors=color,explode=expload_value,autopct='%0.1f%%',textprops=textProp,wedgeprops=wedgeprops)
plt.legend(classes, title="Class Names", loc="best")
plt.show()
```
