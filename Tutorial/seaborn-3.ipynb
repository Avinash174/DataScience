```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.scatterplot(x='age',y='fare',data=df)
```

```python
sns.scatterplot(x='age',y='fare',data=df,hue='alive')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',data=df,hue='class')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',style='class',data=df,hue='alive')
plt.title('Titabic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',style='class',data=df,hue='alive',palette='inferno')
plt.title('Titabic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',style='class',data=df,hue='alive',palette='gist_rainbow',alpha=0.5)
plt.title('Titabic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',style='class',data=df,hue='alive',palette='inferno')
sns.lineplot(x='age',y='fare',data=df,color='red')
plt.title('Titabic Data Analysis')
```
