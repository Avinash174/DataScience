```python
import matplotlib.pyplot as plt
import  seaborn as sns
import numpy as np
import pandas as pd
```

```python
plt.style.use('dark_background')
```

```python
marks_50_student=np.random.randint(0,100,(50))
marks_50_student
```

```python
plt.hist(marks_50_student)
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins)

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='orange')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='red',orientation='horizontal')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='red',orientation='horizontal')

plt.yticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='yellow',orientation='vertical',rwidth=0.6)

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='red',histtype='step')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='red',histtype='barstacked')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='red',histtype='bar')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))

plt.hist(marks_50_student,bins=bins,color='red',histtype='stepfilled')

plt.xticks(np.arange(0,100,5))
plt.show()
```

```python
marks_20_student1=np.random.randint(0,100,(50))
marks_20_student2=np.random.randint(0,100,(50))
```

```python
bins=np.arange(0,100,5)
plt.figure(figsize=(6,6))
plt.hist([marks_20_student1,marks_20_student2],bins=bins,color=['orange','blue'])
plt.xticks=(np.arange(0,100,5))
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Of Students Of 2 Classes')
plt.show()
```
