```python
# merge function
```

```python
data={
    'Roll No.':[1,2,3,4,5],
    'Physics':[23,34,12,45,65]

}

df=pd.DataFrame(data)
print(df)
```

```python
data1={
    'Roll No':[1,2,3,4,5],
    'Chemistry':[34,53,26,11,39]
}

df1=pd.DataFrame(data1)
print(df1)
```

```python
pd.merge(df,df1)
```

```python
df3=pd.DataFrame({'Roll_No':[1,2,3,5,6],'Physics':[11,1,2,3,4]},)
df4=pd.DataFrame({'Roll_No':[1,2,3,4,5],
    'Chemistry':[34,53,26,11,39]})
print(df3)
print()
print(df4)
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,how='left')
```

```python
pd.merge(df3,df4,how='right')
```

```python
pd.merge(df3,df4,how='outer')
```

```python
pd.merge(df3,df4,on='Roll_No')
```
