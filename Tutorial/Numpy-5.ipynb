```python
x_sin=np.arange(0,2 * np.pi,0.1)
y_sin=np.sin(x_sin)
print(y_sin)
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_sin,y_sin)
plt.title('Sin Curve')
plt.show()
```

```python
x_cos=np.arange(0,2 * np.pi,0.1)
y_cos=np.cos(x_cos)
print(y_cos)

plt.figure(figsize=(6,6))
plt.plot(x_cos,y_cos)
plt.title('Cos Curve')
plt.show()
```

```python
x_tan=np.arange(0,2 * np.pi,0.1)
y_tan=np.tan(x_cos)
print(y_tan)

plt.figure(figsize=(6,6))
plt.plot(x_tan,y_tan)
plt.title('Tan Curve')
plt.show()
```

```python
x_cot=np.arange(0,2 * np.pi,0.1)
y_cot=1/np.tan(x_cos)
print(y_tan)

plt.figure(figsize=(6,6))
plt.plot(x_cot,y_cot)
plt.title('Cot Curve')
plt.show()
```

```python
# Sub Plot
```

```python
plt.figure(figsize=(6,6))
plt.plot(x_cot,y_cot)
plt.subplot(2,2,1)
plt.title('Cot Curve')
plt.show()
```
