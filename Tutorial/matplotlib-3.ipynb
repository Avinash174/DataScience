```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
subject=['Math','Science','History','Computer','Marathi']
marks=[30,20,60,50,90]
```

```python
plt.bar(subject,marks)
plt.show()
```

```python
plt.bar(subject,marks,color='green')
plt.show()
```

```python
colors=['red','blue','brown','green','purple']
plt.bar(subject,marks,color=colors)
plt.show()
```

```python
plt.bar(subject,marks,color=colors,width=0.6)
plt.show()
```

```python
plt.bar(subject,marks,color=colors,width=0.8,edgecolor='black')
plt.show()
```

```python
plt.bar(subject,marks,color=colors,width=0.8,edgecolor='black',linewidth=5)
plt.show()
```

```python
plt.bar(subject,marks,color=colors,width=0.8,edgecolor='black',linewidth=3,linestyle='--')
plt.show()
```

```python
plt.barh(subject,marks,color=colors)
plt.show()
```

```python
subject=['Maths','science','social science','computer','Hindi']
marks1=[90,40,12,100,50]
marks2=[30,60,40,10,20]
```

```python
plt.figure(figsize=(8,8))

plt.bar(subject,marks1)
plt.bar(subject,marks2)

plt.xlabel('subject')
plt.ylabel('marks')
plt.show()
```

```python
subject_len=np.arange(len(subject))
width=0.4
```

```python
plt.figure(figsize=(8,8))

plt.bar(subject_len,marks1,width=width)
plt.bar(subject_len+width,marks1,width=width)

plt.xlabel('Subject')
plt.ylabel('Marks')

plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.bar(subject_len,marks1,width=width,color=colors)
plt.bar(subject_len+width,marks1,width=width,color=colors,alpha=0.5)

plt.xlabel('Subject')
plt.ylabel('Marks')

plt.show()
```
