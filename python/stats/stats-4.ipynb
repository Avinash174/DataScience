A survey cliams that in a math test female student tend to score fewer marks than the average marks of 75 out of 100 consider a sample of 24 female students and perform a hypothesis test to check the claim with 90 % confidence

```python
df_female_score.head(2)
```

```python
# consider a list of math score of female students from the data
math_marks=list(df_female_score['math score'])
#calculate sample mean
sample_avg=np.mean(math_marks)
# claculation smaple of standard deviation
sample_std=statistics.stdev(math_marks)
#smaple length 
n=len(math_marks)
# degree of freedem for 1 sample t - test
print('degrees of freedom:',n-1)
```

The null and alternative hypothesis is:
Ho:u>=75
H1:u<75

Here alpha=0.1 and degree of freedem =23 for a one hand tailed test

```python
t_val=round(stats.t.isf(q=0.1,df=23),2)
print('Critical value for one-tailed t-test:',t_val)
```

```python
t_stat,p_val=stats.ttest_1samp(a=math_marks,popmean=75)
t_stat,p_val
```
