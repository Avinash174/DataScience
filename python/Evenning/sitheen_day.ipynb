```python
import pandas as pd

data={
    "First Name":["ross","avinash","monica","jay","veeru","veer"],
    "Last Name":["tribbute","magar","patil","jadhav","lakade","wagh"],
    "Salary":["20000","10000","4000","70000","45000","34900"],
    "EEID":['E1',"E2","E3","E4","E5","E6"]
}
df=pd.DataFrame(data)
df
```

```python
df["Full Name"]=df['First Name']+df['Last Name']
df
```

```python
df["Full Name"]=df['First Name']+" "+df['Last Name'] # spacing 
df
```

```python
# Calculate bonus and assign to both the column and a variable
df["Give Bonus 15 %"] = df["Salary"] * 0.15

```

```python
df["Full Name"]=df['First Name'].str.upper()+" "+df['Last Name'].str.lower()
 # upper lower lower
df
```

```python
df["Full Name"]=df['First Name'].str.capitalize()+" "+df['Last Name'].str.capitalize()
df
```
