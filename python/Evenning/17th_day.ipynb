```python
import pandas as pd 
```

```python
data1={
    "Emp_id":["E01","E02","E03","E04","E05","E06"],
    "Name":["Ram","Sham","Avinash","Rahul","Ravi","Kunnal"],
    "Age":[34,90,10,75,10,22]
}

data2={
    "Emp_id":["E01","E02","E03","E04","E05"],
    "Salary":[4500,2344,1880,7499,8460]
}

df4=pd.DataFrame(data1)
df5=pd.DataFrame(data2)
```

```python
print(df4)
print()
print(df5)
```

```python
pd.merge(df4,df5,on="Emp_id") # join the basis of emp id
```

```python
pd.merge(left=df4,right=df5,on="Emp_id",how="left")
#left join 
```

```python
# 28-12-2024
pd.merge(left=df4,right=df5,on="Emp_id",how="right") #right join
```

```python
data3={

    "Emp_id":["E01","E02","E03","E04","E05","E06"],
    "Name":["Avinash","Ram","Ravi","Rahul","Akash","Ronak"],
    "Age":[34,90,10,75,10,22],
    "Salary":[4500,2344,1880,7499,8460,4590]
}

data4={
    "Emp_id":["E07","E08","E09","E010","E011","E012"],
    "Name":["Rohit","Sham","Avi","Raj","Ravindra","Kunnal"],
    "Age":[34,90,10,75,10,22],
    "Salary":[5000,2344,8080,4099,1460,3590]
}

df3=pd.DataFrame(data3)
df5=pd.DataFrame(data4)

print(df3)
print()
print(df5)
```

```python
# campare dataframe

dict={
    "Fruits":["mango","apples","banana","papaya"],
    "Price":[90,80,60,50],
    "Quantity":[15,10,11,2]
}
df=pd.DataFrame(dict)
print(df)
```

```python
# making copy of dataFrame

df2=df.copy()
print(df2)
print()
print(df)
```

```python
df2.loc[0,"Price"]=123
df2.loc[1,"Price"]=12
df2.loc[3,"Price"]=1
```

```python
df2.loc[0,"Quantity"]=13
df2.loc[1,"Quantity"]=3
df2.loc[3,"Quantity"]=10
```
