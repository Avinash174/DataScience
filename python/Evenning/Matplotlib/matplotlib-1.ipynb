```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
x=[1,2,3]
y=[9,3,1]

plt.plot(x,y,'b',label='Line One',linewidth=5)
plt.xlabel("x axis")
plt.ylabel("Y axis")
plt.title('Sample Plot')
plt.legend()
plt.show()
```

```python
day=['day1','day2','day3','day4','day5']
vivo=[4000,6000,1000,2000,100]
oppo=[15000,20000,600,1000,400]
```

```python
plt.plot(day,vivo,oppo,marker="*",color="blue")
plt.xlabel("Days")
plt.ylabel("sales")
plt.title("line plot")
plt.grid(visible=True,color='grey')
plt.show()
```

```python
plt.plot(day,oppo,marker="*",label='vivo')
plt.plot(day,vivo,marker="*",label='oppo')
plt.xlabel("Days")
plt.ylabel("sales")
# plt.title("line plot")
plt.legend()
plt.grid(visible=True,color='grey')
plt.show()
```

```python
plt.bar(x=day,height=vivo)
plt.show()
```

```python
index=np.arange(4)
plt.bar(x=index,height=vivo,label='vivo')
plt.bar(x=index,height=oppo,bottom=vivo,label='oppo')
plt.xticks(index,day)
plt.show()
```

```python
plt.barh(x=day,height=vivo)
plt.show()
```

```python
index=np.arange(4)
plt.bar(x=index,height=vivo,width=0.35,label='vivo')
plt.bar(x=index + 0.35,height=oppo,width=0.35,label='oppo')
plt.legend()
plt.show()
```
