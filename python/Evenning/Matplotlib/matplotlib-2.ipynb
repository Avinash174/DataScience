```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
import seaborn as sns 
```

```python
df=sns.load_dataset('titanic')
df
```

```python
plt.pie(x=df['class'].value_counts(),labels=df['class'].value_counts().index,autopct='%.2f%%')
plt.show()
```

```python
expload=[0,0,0.1]
plt.pie(x=df['class'].value_counts(),labels=df['class'].value_counts().index,explode=expload,autopct='%0.2f%%')
plt.show()
```

```python
expload=[0,0,0.1]
color=['red','orange','purple']
plt.pie(x=df['class'].value_counts(),labels=df['class'].value_counts().index,explode=expload,autopct='%0.2f%%',colors=color)
plt.show()
```

```python

color=['red','orange','purple']
plt.pie(x=df['class'].value_counts(),labels=df['class'].value_counts().index,autopct='%0.2f%%',colors=color)
circle=plt.Circle(xy=[0,0],radius=0.35,color='white')
plt.gca().add_artist(circle)
plt.show()
```

```python

color=['red','orange','purple']
plt.figure(figsize=(8,8))
plt.pie(x=df['sex'].value_counts(),labels=df['sex'].value_counts().index,autopct='%0.2f%%',colors=color)
circle=plt.Circle(xy=[0,0],radius=0.35,color='white')
plt.gca().add_artist(circle)
plt.show()
```
