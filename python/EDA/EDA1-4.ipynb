```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df
```

```python
#standard scaler

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler_data=scaler.fit_transform(df[['sepal_width','sepal_length']])
scaler_data
```

```python
# Min Max Scaler

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler_data=scaler.fit_transform(df[['sepal_width','sepal_length']])
scaler_data
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sepal_width'], y=df['sepal_length'])  # Correct column names
plt.show()
```

```python
np.log1p(df['sepal_length'])
```

```python
sns.histplot(df['sepal_length'])
```

Log Transform

```python
col_log=df['col_log']=np.log(df['sepal_width'])
col_log
```

Box-Cox Transformatio

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='box-cox')
trans_col = pt.fit_transform(df[['sepal_length', 'sepal_width']])
trans_col
```

```python
sns.histplot(df['col_log'])
```

Log1p Transformation

```python
col_log1p=df['col_log1p']=np.log1p(df['sepal_width'])
col_log1p
```

```python
sns.histplot(df['col_log1p'])
```

yeo-johnson

```python
yeo = PowerTransformer(method='yeo-johnson')
trans_col = yeo.fit_transform(df[['sepal_length']])
trans_col
```

```python
sns.histplot(trans_col)
```
